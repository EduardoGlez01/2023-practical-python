## Multipage Tabular Scrape

- <a href="https://apps.health.ny.gov/pubdoh/professionals/doctors/conduct/factions/AllRecordsAction.action">On this site</a>, scrape all doctors whose last names begin with "P".
- Export the content into a CSV file called ```md_P.csv```.


````python
## create cells as needed
url = "https://apps.health.ny.gov/pubdoh/professionals/doctors/conduct/factions/AllRecordsAction.action"
response = requests.get(url)
response.status_code
````

````python
## Check type
type(response.text)
````

````python
## Checking what value contains the elements
df_p = pd.read_html(response.text)[1]
df_p
````

````python
base_url = "https://apps.health.ny.gov/pubdoh/professionals/doctors/conduct/factions/AllRecordsAction.action?d-49653-p="
````

````python
##For loop to get all the URLS
docs_url = []
for url_number in range(1,439):
    docs_url.append(f"{base_url}{url_number}.html")
````

````python
docs_url
````

````python
##IMPORT FUNCTIONS
from random import randrange
import time
````

````python
## Scraping data

busted_links = []
df_all = []
total_links = len(docs_url)
counter = 1

for url in docs_url:
    print(f"Scraping {counter} of {total_links}")
    counter += 1
    response = requests.get (url)
    try: 
        df = pd.read_html(response.text)[1]
    except:
        print(f"oh no! {url} returned {response.status_code}")
        busted_links.append(url)
    else: 
        df_all.append(df)
        print(df)
    snoozer = randrange(2,5)
    print(f"snoozing for {snoozer} seconds before next scrape")
    time.sleep(snoozer)
    
print ("done scraping all provided links")
````

````python
df_all[1]
````

````python
## Concat all the urls in one dataframe
df = pd.concat(df_all, ignore_index = True)
df
````

````python
type(df)
````

````python
## Filter the values that start with P

df_only_p = df[df['Physician Last Name'].str.startswith('P')]
print(df_only_p)
````

````python
## Export to CSV

df_only_p.to_csv("md_P.csv", index = False, encoding = "UTF-8")
````
